```python
import torch
import torch.nn as nn
from torchvision import models

# ConvNeXt 모델 불러오기 (pretrained=True로 사전 학습된 가중치 사용)
model = models.convnext_tiny(pretrained=True)

# Feature Extractor로 사용하기 위해 마지막 분류 계층 제거
class ConvNeXtFeatureExtractor(nn.Module):
    def __init__(self, original_model):
        super(ConvNeXtFeatureExtractor, self).__init__()
        # 마지막 분류 레이어를 제외한 모든 레이어를 유지
        self.features = nn.Sequential(*list(original_model.children())[:-1])  # 분류기 제거

    def forward(self, x):
        # 중간 feature map 추출
        x = self.features(x)
        return x

# Feature Extractor 모델 정의
feature_extractor = ConvNeXtFeatureExtractor(model)

# 모델을 평가 모드로 설정
feature_extractor.eval()

large_image = torch.randn(1, 3, 224, 224)

# Feature 추출
with torch.no_grad():
    large_features = feature_extractor(large_image)

# Feature 출력 크기 확인
print("Extracted Feature Shape for 224x224 input:", large_features.shape)
```
